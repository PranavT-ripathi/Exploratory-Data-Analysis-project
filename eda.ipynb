```python
ax=sns.countplot(x='Gender',data=df,hue='Marital_Status')   #hue adds one more dimension within an existing dimension

for bars in ax.containers: #iterates over all the bars in the plot
    ax.bar_label(bars)     #add labels to each bar (typically heights)
```

```python
gender_amount=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
#similar to select gender,sum(amount) from df group by gender 
ax=sns.barplot(x='Gender',y='Amount',data=gender_amount)

for bars in ax.containers:
    ax.bar_label(bars)
```

## the number of females is more and they have spent more money


# age

```python
ax=sns.countplot(x='Age Group', data=df)
```

```python
ax=sns.countplot(x='Age Group',data=df,hue='Gender')
```

```python
age_amount=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
ax=sns.barplot(x='Age Group',y='Amount',data=age_amount)
#for bars in ax.containers:
    #ax.bar_label(bars)
```

### people between age 26-35 have spent the most amount and most of them are females

# state

```python
state_amount=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(5)
sns.barplot(x='State',y='Amount',data=state_amount)
```

## people from uttar Pradesh have spent the most amount of money

# marital status

```python
ax=sns.countplot(x='Marital_Status',data=df,hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
marital_amount=df.groupby(['Marital_Status'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Marital_Status',y='Amount',data=marital_amount)
```

```python
marital_gender_amount=df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Marital_Status',y='Amount',data=marital_gender_amount,hue='Gender')
```

### most amount is spent by unmaried females

#  occupation

```python
occupation_amount=df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(5)
sns.barplot(x='Occupation',y='Amount',data=occupation_amount)
```

### most amount is spent by people working in IT

# product category


```python
df.columns
```

```python
category_orders=df.groupby(['Product_Category'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(3)
sns.barplot(x='Product_Category',y='Orders',data=category_orders)
```

### most of the orders are placed in clothes and apparel category

```python
category_amount=df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(3)
sns.barplot(x='Product_Category',y='Amount',data=category_amount)
```

### most amount has been spent on food
